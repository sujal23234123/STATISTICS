```python
import numpy as np

matrix1=np.array([[4,6,3],[5,7,8]])
matrix2=np.array([[8,6,3],[8,4,9]])

print("\nfirst matrix is :",matrix1)
print("\nsecond matrix is :",matrix2)


print(np.cov(matrix1))
print(np.cov(matrix2))
print(np.corrcoef(matrix1))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

study_hour=[4,5,3,6,8,4,6,7,7,8,4,7,6]
test_scores=[6,7,8,9,4,6,8,7,4,6,7,8,9]

correlation,_=pearsonr(study_hour,test_scores)
print("correlation cofficient :",round(correlation,2))

plt.bar(study_hour,test_scores)
plt.title("study hour vs test score")
plt.xlabel("study hours")
plt.ylabel("test scores")
plt.show()
```
