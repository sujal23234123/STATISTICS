#TO CALCULATE IQR--

1-SORT THE DATASET

2-CALCULATE Q1(25%) and Q3(75%) using percentile

3-calculate IQR(Q3-Q1)

4-CALCULATE LOWER_FENCE(Q1-(1.5*IQR))

5-CALCULATE HIGHER_FENCE(Q3+(1.5*IQR))

6-CHECK WITH BOXPLOT

```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


dataset=[12,23,43,32,54,65,43,65,76,54,67,45,34,65,24,75,68,65,86,42,6,102]
sorted_dataset=sorted(dataset)
print("The sorted dataset of yours  :",sorted_dataset)

#Q1

Q1=np.percentile(dataset,[25]) 
print("The Q1 value is :",Q1)

#Q3

Q3=np.percentile(dataset,[75])
print("The Q3 value is :",Q3)

#IQR
IQR=Q3-Q1
print("The IQR is :",IQR)

#lower_fence

lower_fence=Q1-(1.5*IQR)
print("The lower_fence value is :",lower_fence)

#higher_fence

higher_fence=Q3+(1.5*IQR)
print("The higher_fence value is :",higher_fence)

#conclusion

print(f"the value range from {lower_fence} to {higher_fence} if it is outside of this range then it is an outlier")

sns.boxplot(dataset)

```
