```python
#normal data distribution
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

fig,axs=plt.subplots(3,2,figsize=(14,12))
fig.suptitle("Various Histograms of various data ",fontsize=16)

data_normal=np.random.normal(loc=0,scale=1,size=1000)
sns.histplot(data_normal,kde=True,ax=axs[0,0],color="blue",stat="density",bins=30)
axs[0,0].set_title("normal data distribution")
axs[0,0].grid(True)

#skewed data distribution

data_right_skewed=np.random.exponential(scale=1,size=1000)
data_left_skewed=np.random.exponential(scale=1,size=1000)
sns.histplot(data_right_skewed,ax=axs[0,1],color="blue",kde=True,bins=10,stat="density",label="right skewed")


sns.histplot(data_left_skewed,ax=axs[0,1],color="red",kde=True,bins=10,stat="density",label="left skewed")

axs[0,1].set_title("skewed data distribution")
axs[0,1].grid(True)
axs[0,1].legend()

#uniform data distrubution

data_uniform=np.random.uniform(low=0,high=1,size=1000)
sns.histplot(data_uniform,ax=axs[1,0],color="green",kde=True,stat="density",bins=20)
axs[1,0].grid(True)
axs[1,0].set_title("Uniform data distribution")

#bimodel distribution

data_bimodel=np.concatenate([np.random.normal(loc=2,scale=1,size=500),
                             np.random.normal(loc=2,scale=1,size=500)])

sns.histplot(data_bimodel,ax=axs[1,1],color="orange",kde=True,stat="density",bins=30)
axs[1,1].set_title=("bimodel data distribution")
axs[1,1].grid(True)

#multimodel distribution

data_multimodel=np.concatenate([
    np.random.normal(loc=6,scale=1.5,size=500),
    np.random.normal(loc=0,scale=1,size=500),
    np.random.normal(loc=6,scale=1,size=500)
])
sns.histplot(data_multimodel,kde=True,ax=axs[2,0],color="brown",stat="density",bins=30)
axs[2,0].set_title("multimodel distribution")
axs[2,0].grid(True)

axs[2,1].axis("off")


```
